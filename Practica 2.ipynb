Para esta práctica se ha seleccionado una base de datos proviniente de Kaggle.

Esta es una base de datos que nos muestra los estudios de algunos pacientes que tienen diabetes y los que no tienen.

La basa de datos contiene las siguientes variables:

*	Número de embarazos
*	Nivel de glucosa en sangre
*	Presión arterial
*	Grosor de la piel
*	Nivel de insulina
*	Indice de masa corporal (BMI)
*	Función de pedigrí de diabetes
*	Edad
*	Resultado de la diabetes (0 para no diabético y 1 para diabético).

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
#Base de datos
df = pd.read_csv("diabetes.csv")
df.head(10)
```

```
#Tipos de datos
df.info()
```

```
#Numero de columnas y numero de registros
df.dtypes
df.shape
```

```
#Estadistica descriptiva de la base de datos
df.describe()
```

```
#Obtenemos los graficos de densidad por cada variable
a=5
b=2
it=1
for i in df:
    plt.subplot(a,b,it)
    sns.kdeplot(x = df.loc[:,i], data = df)
    plt.grid()
    it+=1
plt.tight_layout()
plt.show()
```

```
#Obtenemos la correlacion de las variables
matriz_correlacion = df.corr()
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```

Observamos que las variables que tienen mas correlacion positiva es la edad y los embarazos
